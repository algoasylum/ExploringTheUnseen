```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set(style='whitegrid')
sns.set_palette('bright')
import warnings
warnings.filterwarnings('ignore')
from scipy import fftpack
```

```python
in_array = np.linspace(-np.pi, np.pi, 24) 
out_array = np.sin(in_array) 
print(len(in_array))  
time = np.linspace(0,24,24)
print(len(time))
print("in_array : ", in_array) 
print("\nout_array : ", out_array)   
# red for numpy.sin() 
plt.plot(time, out_array, color = 'red', marker = "o") 
plt.title("numpy.sin()") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show() 
```

```python
dailyx = np.tile(out_array,365*10)
print(dailyx.shape)
```

```python
time=np.linspace(0,87600,87600)
plt.plot(time[:100], dailyx[:100], color = 'red')
```

```python
in_array = np.linspace(-np.pi, np.pi, 8760) 
yearlyx = np.sin(in_array) 
print(len(in_array))  
time = np.linspace(0,8760,8760)
print(len(time))
#print("in_array : ", in_array) 
#print("\nout_array : ", out_array)   
# red for numpy.sin() 
plt.plot(time, yearlyx, color = 'red') 
plt.title("numpy.sin()") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show() 
```

```python
Y = np.fft.fft(dailyx)
freq = np.fft.fftfreq(len(dailyx))
Y.real
```

```python
plt.plot( freq, np.abs(Y) )
```

```python
def freq(df):
    N = df.shape[0]
    y = df
    f = fftpack.fftfreq(len(y))
    yf = fftpack.fft(y)
    print(yf)
    freq_ampli = pd.DataFrame(zip(np.abs(f[:N//2+1]), 2.0 / N * np.abs(yf[:N//2+1])),columns=['Freq','Ampli'])
    return freq_ampli 
freq_ampli1 = freq(dailyx)
```

```python
fdai = freq_ampli1.sort_values(by=['Ampli'],ascending=False)
fdai
```
