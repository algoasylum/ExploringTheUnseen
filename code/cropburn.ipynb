```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import plotly.express as px
import seaborn as sns
sns.set(style='whitegrid')
sns.set_palette('bright')
import warnings
warnings.filterwarnings('ignore')
```

```python
city = ['patiala', 
        'hisar', 
        #'jind',
        'panipat', 
        'sukhdev', 
        #'stadium',
        'noida']
```

```python
df = pd.read_csv('data/cropburn.csv')
df = df[df.city.isin(city)]
df.date = pd.to_datetime(df.date,format = '%d-%m-%Y')
df20 = df[df['date'].dt.year == 2019]
df20 = df20[(df20['date'].dt.month == 10) | (df20['date'].dt.month == 11) | (df20['date'].dt.month == 12)]
df20 = df20.reset_index()
df20.city.unique()
```

```python
fig = plt.subplots(1,1,figsize=(14,8))
sns.lineplot(df20['date'],df20['pm25'],hue=df20.city)
```

```python
fig = plt.subplots(1,1,figsize=(14,8))
sns.lineplot(df20['date'],df20['pm10'],hue=df20.city)
```

```python
fig = plt.subplots(1,1,figsize=(14,8))
sns.lineplot(df20['date'],df20['no2'],hue=df20.city)
```

```python
fig = plt.subplots(1,1,figsize=(14,8))
sns.lineplot(df20['date'],df20['so2'],hue=df20.city)
```

```python
fig,a = plt.subplots(7,3,figsize=(14,18))

for l,i in enumerate(city):
    dfc = df[df.city.isin([i])]
    dfc =  dfc[dfc.date.dt.year == 2019]
    dfc =  dfc[dfc.date.dt.month == 10]
    sns.distplot(dfc['pm25'],bins=50,ax = a[l][0])
    a[l][0].set_title(i,c='red')
for l,i in enumerate(city):
    dfc = df[df.city.isin([i])]
    dfc =  dfc[dfc.date.dt.year == 2019]
    dfc =  dfc[dfc.date.dt.month == 11]
    sns.distplot(dfc['pm25'],bins=50,ax = a[l][1])
    a[l][1].set_title(i,c='red')
for l,i in enumerate(city):
    dfc = df[df.city.isin([i])]
    dfc =  dfc[dfc.date.dt.year == 2019]
    dfc =  dfc[dfc.date.dt.month == 12]
    sns.distplot(dfc['pm25'],bins=50,ax = a[l][2])
    a[l][2].set_title(i,c='red')
plt.tight_layout()
plt.show()
```
