```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set(style='whitegrid')
sns.set_palette('bright') 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/2020cropburn.csv')
df['utc'] = pd.to_datetime(df['utc'],format = '%Y-%m-%d')
df
```

```python
gas = df.parameter.unique()
city = df.city.unique()
city = ['Hisar', 'Ludhiana', 'Muzaffarnagar', 'Greater Noida','Patiala', 'Delhi','Jind']
df = df[df.city.isin(city)]
```

```python
dff = df[df.parameter.isin(['pm25'])]
fig = plt.subplots(1,1,figsize=(14,8))
sns.lineplot(dff['utc'],dff['value'],hue=dff.city)
plt.show()
```

```python
dff = df[df.parameter.isin(['pm10'])]
fig = plt.subplots(1,1,figsize=(14,8))
sns.lineplot(dff['utc'],dff['value'],hue=dff.city)
plt.show()
```
