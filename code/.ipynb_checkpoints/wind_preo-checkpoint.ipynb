```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv(f'data/wpanipat.csv')
df2 = pd.read_csv(f'data/wpatiala.csv')
df3= pd.read_csv(f'data/whisar.csv')
df4= pd.read_csv(f'data/wnoida.csv')
df5= pd.read_csv(f'data/wusembassy.csv')
df6= pd.read_csv(f'data/wjind.csv')
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6])
df.columns = ['city','date','value']
df
```

```python
df.date = pd.to_datetime(df.date,format='%d-%b-%y')
df.to_csv(f'data/uwind.csv')
```
