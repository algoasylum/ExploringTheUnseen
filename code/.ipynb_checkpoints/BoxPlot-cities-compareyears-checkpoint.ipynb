## Summary Statistics

This notebook compares the change in pollution for different cities across different years. The data is read from `year_7cities.csv`

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

plt.style.use('seaborn-bright')
```

```python
year1 = 2019
year2 = 2020
df_y1 = pd.read_csv(f'data/{year1}_7cities.csv')
df_y2 = pd.read_csv(f'data/{year2}_7cities.csv')
```

### Clean up:
- Make the Date column of the right type, filter the years, add month numbers.
- Remove min, max, variance
- rename 'median' do something that doesn't clash with an existing name!!

```python
df_y1.Date = pd.to_datetime(df_y1.Date, format = '%d/%m/%Y')
df_y1 = df_y1[df_y1['Date'].dt.year == year1]
df_y1['month_num'] = pd.DatetimeIndex(df_y1['Date']).month
df_y1.rename(columns = {'median':'value'}, inplace=True)

df_y2.Date = pd.to_datetime(df_y2.Date, format = '%d/%m/%Y')
df_y2 = df_y2[df_y2['Date'].dt.year == year2]
df_y2['month_num'] = pd.DatetimeIndex(df_y2['Date']).month
df_y2.rename(columns = {'median':'value'}, inplace=True)
```

```python
df_y1.drop(['Unnamed: 0', 'count', 'min', 'max', 'variance', 'Country'], axis=1, inplace=True)
df_y2.drop(['Unnamed: 0', 'count', 'min', 'max', 'variance', 'Country'], axis=1, inplace=True)
```

```python
df_y1.head(5)
```

```python
#just to make this generic
cities = df_y1.City.unique()
pollutants = df_y1.Specie.unique()

print(cities)
print(pollutants)

## 24-hour means unless otherwise specified. Using 8-hour as a proxy when unavailable
pollutant_who = {}
pollutant_who['co']   = 0
pollutant_who['no2']  = 0
pollutant_who['o3']   = 100 ##  8-hour mean
pollutant_who['pm10'] = 50
pollutant_who['pm25'] = 25
pollutant_who['so2']  = 20  

## annual in 'industrial, residential, rural, others' areas
pollutant_naaqs = {}
pollutant_naaqs['co']   = 2   ##  8-hour mean
pollutant_naaqs['no2']  = 80
pollutant_naaqs['o3']   = 100 ##  8-hour mean
pollutant_naaqs['pm10'] = 100
pollutant_naaqs['pm25'] = 60
pollutant_naaqs['so2']  = 80  
```

---

I'm really, really partial to making grids of plots myself!

```python
combo = pd.concat([df_y1, df_y2], sort=False)
combo['year'] = pd.DatetimeIndex(combo['Date']).year
combo.dtypes
```

```python
# let's plot the data for one city to see what it looks like
fig, ax = plt.subplots(1, len(pollutants), figsize=(24,6))
df = df_y2
city = 'Delhi'
for i, p in enumerate(pollutants):
    d = df[df.Specie == p]     
    a = ax[i]
    d[d.City == city].boxplot(column='value', by='month_num', ax=a)
    a.set_title(p, color='blue')
    a.set(xlabel='', ylabel='')
    if pollutant_who[p]:
        a.axhline(y=pollutant_who[p], c='g')
    if pollutant_naaqs[p]:
        a.axhline(y=pollutant_naaqs[p], c='r')
plt.suptitle(f'{city}', fontsize=24, color='b', y=-0.01)
plt.show()
plt.savefig(f'images/boxplot-{city}-poll-{year2}.png')
plt.close()
```

```python
# same, but with seaborn
fig, ax = plt.subplots(1, len(pollutants), figsize=(24,6))
df = df_y2
city = 'Delhi'

for i, p in enumerate(pollutants):
    d = df[df.Specie == p]     
    a = ax[i]
    sns.set_style("whitegrid")
    sns.boxplot(x=d.month_num, y=d.value, color='lightblue', ax=a)
    #d[d.City == city].boxplot(column='value', by='month_num', ax=a)
    a.set_title(p, color='blue')
    a.set(xlabel='', ylabel='')
    if pollutant_who[p]:
        a.axhline(y=pollutant_who[p], c='g')
    if pollutant_naaqs[p]:
        a.axhline(y=pollutant_naaqs[p], c='r')
plt.suptitle(f'{city}', fontsize=24, color='b', y=-0.01)
plt.savefig(f'images/boxplot-{city}-poll-{year2}.png')
plt.show()
```

```python
# only three pollutants:
fig, ax = plt.subplots(1, 3, figsize=(24,6))
city = 'Mumbai'
df = combo[(combo.City == city)]

for i, p in enumerate(['co', 'pm10', 'pm25']):
    d = df[df.Specie == p]     
    a = ax[i]
    sns.set_style("whitegrid")
    bp = sns.boxplot(x=d.month_num, y=d.value, hue=d.year, palette='Set3', ax=a)
    a.set(xlabel='', ylabel='')
    bp.legend_.remove()
    #d[d.City == city].boxplot(column='value', by='month_num', ax=a)
    a.set_title(p, color='blue')
    if pollutant_who[p]:
        a.axhline(y=pollutant_who[p], c='g')
    if pollutant_naaqs[p]:
        a.axhline(y=pollutant_naaqs[p], c='r')
plt.suptitle(f'{city}', fontsize=24, color='b', y=-0.001)
plt.savefig(f'images/boxplot-{city}-poll-{year1}-{year2}-3p.png', bbox_inches='tight')
plt.show()
```

```python
## now do it for all the cities..

for city in cities:
    fig, ax = plt.subplots(1, len(pollutants), figsize=(24,6))
    df = combo[(combo.City == city)]
    for i, p in enumerate(pollutants):
        d = df[df.Specie == p]     
        a = ax[i]
        sns.set_style("whitegrid")
        bp = sns.boxplot(x=d.month_num, y=d.value, hue=d.year, palette='Set3', ax=a)
        bp.legend_.remove()
        a.set_title(p, color='blue')
        a.set(xlabel='', ylabel='')
        if pollutant_who[p]:
            a.axhline(y=pollutant_who[p], c='g')
        if pollutant_naaqs[p]:
            a.axhline(y=pollutant_naaqs[p], c='r')
    plt.suptitle(f'{city}', fontsize=24, color='b', y=-0.001)
    plt.savefig(f'images/boxplot-{city}-poll-{year1}-{year2}.png', bbox_inches='tight')
    plt.close()
```

```python
## ... and all pollutants

for p in pollutants:
    fig, ax = plt.subplots(1, len(cities), figsize=(24,6))
    df = combo[(combo.Specie == p)]
    for i, c in enumerate(cities):
        d = df[df.City == c]     
        a = ax[i]
        sns.set_style("whitegrid")
        bp = sns.boxplot(x=d.month_num, y=d.value, hue=d.year, palette='Set3', ax=a)
        bp.legend_.remove()
        a.set_title(c, color='blue')
        a.set(xlabel='', ylabel='')
        if pollutant_who[p]:
            a.axhline(y=pollutant_who[p], c='g')
        if pollutant_naaqs[p]:
            a.axhline(y=pollutant_naaqs[p], c='r')
    fig.suptitle(f'{p}', fontsize=24, color='b', y=-0.001)
    plt.savefig(f'images/boxplot-{p}-city-{year1}-{year2}.png', bbox_inches='tight')
    plt.close()
```

```python
fig, ax = plt.subplots(len(pollutants), len(cities), figsize=(24,36), sharey='row')
plt.subplots_adjust(hspace=0.3) ## the hspace value is a fraction of the average axes height

# For each pollutant, and each city, create a box plot, and put it in the correct (pollutant, city) location
for i, p in enumerate(pollutants):
    d = df[(df.Specie == p)] # extract the pollutant data
    for j, c in enumerate(cities):
        a = ax[i][j] # figure out the axes to use
        d[d.City == c].boxplot(column='median', by='month_num', ax=a)
        if pollutant_who[p]:
            a.axhline(y=pollutant_who[p], c='g')
        if pollutant_naaqs[p]:
            a.axhline(y=pollutant_naaqs[p], c='r')
        a.set_title(c)

for a, r in zip(ax[:,0], pollutants):
    a.set_ylabel(r, rotation=90, fontsize=24, color='b')
    
#plt.title('') # SKK: I need to figure this out. If I uncomment this line, the argument shows up as the title of the last subfigure!
plt.suptitle('')
plt.show()
fig.savefig(f'images/boxplot-pollutants-cities-{year}.png', bbox_inches='tight')
plt.close()
```

```python
fig, ax = plt.subplots(len(pollutants), len(cities), figsize=(24,36), sharey='row')
plt.subplots_adjust(hspace=0.3)

## same as the plot above, but using the daily max, rather than the daily median values for each month
for i, p in enumerate(pollutants):
    d = df[df.Specie == p]
    for j, c in enumerate(cities):
        
        a = ax[i][j]
        d[d.City == c].boxplot(column='max', by='month_num', ax=a)
        a.set_title(c, color='blue')
        if pollutant_who[p]:
            a.axhline(y=pollutant_who[p], c='g')
        if pollutant_naaqs[p]:
            a.axhline(y=pollutant_naaqs[p], c='r')

for a, r in zip(ax[:,0], pollutants):
    a.set_ylabel(r, rotation=90, fontsize=24, color='b')
#plt.title('')
plt.suptitle('')
plt.show()
fig.savefig(f'images/boxplot-pollutants-cities-max-{year}.png', bbox_inches='tight')
plt.close()
```
