```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set(style='whitegrid')
sns.set_palette('bright') 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/2020cropburn.csv')
df.utc = pd.to_datetime(df.utc,format = '%Y-%m-%d')
df.head(3)
```

```python
dff = df[df.parameter == 'pm10']
city=dff.city.unique()
city = ['Hisar', 'Jind', 'Greater Noida','Patiala', 'Delhi']
city
```

```python
fig, ax = plt.subplots(1,5,figsize=(15,5),sharex=True,sharey=True)
z=0
for i,l in enumerate(city):
    df1 = dff[dff.city.isin([l])]
    sns.distplot(df1.value,ax=ax[i],bins=75)
    ax[i].set_title(l, fontsize=17,c='red')
```

```python
d = df[((df['utc'].dt.month == 10) | (df['utc'].dt.month == 11) | (df['utc'].dt.month == 12)) & (df['utc'].dt.year == 2020)]
d.head()
```

```python
fig, ax = plt.subplots(1,5,figsize=(15,5),sharex=True,sharey=True)
z=0
for i,l in enumerate(city):
    df1 = d[d.city.isin([l])]
    sns.distplot(df1.value,ax=ax[i],bins=75)
    ax[i].set_title(l, fontsize=17,c='red')
```
